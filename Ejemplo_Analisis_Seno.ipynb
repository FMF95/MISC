## ‚öôÔ∏è ¬øQu√© es un an√°lisis seno (o ensayo senoidal)?

#### Es una prueba de vibraci√≥n controlada en la que el excitador (shaker) aplica una se√±al senoidal pura a una estructura, variando la frecuencia a lo largo de un rango definido.
#### El objetivo es observar la respuesta del sistema a esa excitaci√≥n arm√≥nica para identificar sus frecuencias naturales, modos de vibraci√≥n, y posibles resonancias.

---

### üîπEn ingenier√≠a o f√≠sica: An√°lisis senoidal (o an√°lisis de se√±ales senoidales)

Un un ‚Äúensayo senoidal‚Äù (sine test), una t√©cnica com√∫n en ingenier√≠a de vibraciones para caracterizar el comportamiento din√°mico de un sistema o componente. Es el estudio del comportamiento de un sistema cuando se somete a una se√±al senoidal (una onda tipo ‚Äúseno‚Äù).

üëâ <b>Usos</b>: an√°lisis de circuitos el√©ctricos, vibraciones, ac√∫stica, estructuras, din√°mica, etc.

üëâ <b>Ejemplo</b>: aplicar una carga o se√±al tipo seno a una estructura o sistema y observar su respuesta (amplitud, fase, frecuencia natural, resonancia, etc.).

üëâ <b>Objetivo</b>: determinar c√≥mo el sistema responde a est√≠mulos arm√≥nicos, lo que ayuda a predecir su comportamiento ante se√±ales m√°s complejas (mediante an√°lisis de Fourier).

---

üîç Principio b√°sico

Una se√±al senoidal tiene la forma:

$$ x(t) = A\sin(2ùúãùëìt +ùúô) $$

donde:

- $A$: amplitud (nivel de vibraci√≥n),

- $f$: frecuencia,

- $ùúô$ = fase inicial.

Durante el ensayo, se barre la frecuencia $f$ (por ejemplo, de 5 Hz a 2000 Hz), mientras se mantiene una amplitud o nivel de aceleraci√≥n constante (por ejemplo, 0.5 $g$ RMS).

 ---

### üìà Tipos de an√°lisis seno

<b>Sine Sweep (barrido senoidal)</b>:
La frecuencia del excitador se incrementa o disminuye lineal o logar√≠tmicamente dentro de un rango.
‚Üí Sirve para detectar resonancias (picos de respuesta).

<b>Sine Dwell</b>:
Se mantiene la excitaci√≥n a una frecuencia fija (de resonancia) durante un tiempo.
‚Üí Se usa para evaluar la fatiga o el comportamiento estructural en resonancia.

---

### üß© Resultados t√≠picos

Del an√°lisis se obtienen curvas de:

- Respuesta en amplitud (aceleraci√≥n, velocidad o desplazamiento) vs. frecuencia.

- Desfase (fase entre excitaci√≥n y respuesta).

- Estos datos permiten calcular:

- Frecuencias naturales,

- Factores de amortiguamiento,

- Rigidez din√°mica,

- Modos de vibraci√≥n (si se instrumenta en varios puntos).

---

### üß™ Aplicaciones comunes

- Ensayos de certificaci√≥n aeroespacial (MIL-STD-810, DO-160).

- Calibraci√≥n de modelos de elementos finitos (FEA).

- Validaci√≥n estructural o de ensamblajes.

- Verificaci√≥n de equipos electr√≥nicos o mec√°nicos frente a vibraciones.

---

### üí° Diferencia con otros tipos de an√°lisis

| Tipo de ensayo         | Se√±al aplicada               | Uso principal                                                 |
| ---------------------- | ---------------------------- | ------------------------------------------------------------- |
| **Seno (Sine)**        | Onda senoidal pura           | Resonancias y respuesta arm√≥nica                              |
| **Aleatorio (Random)** | Ruido con espectro definido  | Simulaci√≥n de vibraciones reales (como en vuelo o transporte) |
| **Chirp o Sine Sweep** | Seno con frecuencia variable | Identificaci√≥n modal o b√∫squeda de resonancias                |
| **Shock (Choque)**     | Pulso transitorio            | Impactos o eventos bruscos                                    |

### ‚öôÔ∏è 1. Qu√© informaci√≥n se obtiene de un an√°lisis seno

Cuando realizas un ensayo senoidal (sine test), el objetivo es caracterizar din√°micamente el sistema, es decir, entender c√≥mo responde a excitaciones peri√≥dicas.
La informaci√≥n que obtienes se deriva de la relaci√≥n entre la excitaci√≥n (entrada) y la respuesta medida (salida).

#### üîπ Se√±ales medidas

Durante el ensayo, normalmente se miden:

- Aceleraci√≥n de entrada (sensor en el ‚Äúshaker‚Äù o punto de fijaci√≥n).

- Aceleraciones de respuesta (uno o varios aceler√≥metros en la estructura).

A partir de estas se√±ales se obtienen:

| Tipo de dato                                  | Descripci√≥n                                                                 | Qu√© te dice                                                      |
| --------------------------------------------- | --------------------------------------------------------------------------- | ---------------------------------------------------------------- |
| **Amplitud (magnitud)**                       | Aceleraci√≥n, velocidad o desplazamiento de respuesta                        | Cu√°nto vibra el sistema a cada frecuencia                        |
| **Fase**                                      | Diferencia de fase entre entrada y salida                                   | C√≥mo se retrasa o adelanta la respuesta respecto a la excitaci√≥n |
| **Transmisibilidad (Transfer Function, FRF)** | Cociente entre salida y entrada en frecuencia: $H(f) = \frac{Y(f)}{X(f)}$ | Describe la respuesta din√°mica completa                          |
| **Resonancias**                               | Frecuencias donde la amplitud de respuesta es m√°xima                        | Coinciden con las **frecuencias naturales** del sistema          |
| **Amortiguamiento (Œ∂)**                       | Se calcula a partir del ancho de banda de un pico resonante                 | Indica la capacidad del sistema para disipar energ√≠a             |

---


### üß† 2. Interpretaci√≥n de los resultados

Al hacer un barrido de frecuencia (sine sweep), obtienes una curva como esta:

```txt
Amplitud

‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚Üí Frecuencia

```
- El pico indica una frecuencia natural (f‚ÇÅ) del sistema.

- La amplitud m√°xima crece cuando el amortiguamiento es bajo.

#### üìâ b) Curva de fase vs. frecuencia

Esto confirma el comportamiento t√≠pico de un oscilador amortiguado.

### üîß 3. C√≥mo se define e introduce la amplitud (RMS, pico, pico a pico)

En vibraciones, la amplitud de excitaci√≥n puede expresarse de tres formas:

| Tipo                       | S√≠mbolo   | Relaci√≥n                            | Comentario                   |
| -------------------------- | --------- | ----------------------------------- | ---------------------------- |
| **Pico (peak)**            | $A_{pk}$  | ‚Äî                                   | Valor m√°ximo instant√°neo     |
| **Pico a pico (p-p)**      | $A_{pp}$  | $A_{pp} = 2A_{pk}$                  | De cresta a cresta           |
| **RMS (Root Mean Square)** | $A_{rms}$ | $A_{rms} = \frac{A_{pk}}{\sqrt{2}}$ | Valor eficaz (energ√≠a media) |

La amplitud RMS (valor cuadr√°tico medio) representa la energ√≠a equivalente de la se√±al:

$$ A_{rms} ‚Äã= \sqrt{ \frac{1}{T} ‚Äã‚à´_{0}^{T} ‚Äã[A sin(2œÄft)]^{2} dt } ‚Äã= \frac{A}{\sqrt{‚Äã2}}‚Äã $$

üëâ Por eso, los niveles de vibraci√≥n especificados en est√°ndares (como MIL-STD o ISO) se suelen dar en aceleraci√≥n RMS ($g$ RMS), porque:

- Refleja la energ√≠a transmitida, no solo el pico.

- Es m√°s √∫til para comparar se√±ales o definir la severidad del ensayo.

---

### üìä 4. C√≥mo se controla la amplitud en el ensayo

En el sistema de control del shaker:

1. Se programa un perfil de frecuencia (por ejemplo, 5 Hz ‚Üí 2000 Hz en 2 min).

2. Se define un nivel de excitaci√≥n constante (por ejemplo, 0.5 $g$ RMS).

3. El controlador de vibraci√≥n ajusta en tiempo real la salida al amplificador del shaker para mantener esa amplitud RMS medida por el aceler√≥metro de control.

Si el sistema entra en resonancia y la amplitud tiende a subir, el controlador reduce la se√±al para mantener el nivel RMS deseado ‚Äî esto evita da√±os.

---

### üß© 5. Par√°metros derivados importantes

A partir de las curvas de respuesta puedes calcular:

<b>1. Frecuencia natural ($f_{n}$)</b>

Frecuencia en la que ocurre la amplitud m√°xima.

<b>2. Amortiguamiento ($Œ∂$ o $Q$)</b>

Del ancho del pico a ‚àí3 dB respecto al m√°ximo:
‚Äã
$$ Œ∂ = \frac{Œîf‚Äã}{2fn‚Äã} $$

o

$$ Q = \frac{1}{2Œ∂} ‚Äã= \frac{fn‚Äã‚Äã}{Œîf} $$

<b>3. Rigidez din√°mica ($k^{*}$)</b>

Relaciona fuerza y desplazamiento:

$$ k^{‚àó}(œâ) = \frac{F(œâ)‚Äã}{X(œâ)} $$


Permite comparar rigidez aparente del sistema en distintas frecuencias.

<b>4. Identificaci√≥n modal (si hay varios sensores)</b>

Usando las funciones de transferencia entre m√∫ltiples puntos, puedes obtener formas modales experimentales y compararlas con modelos FEA.

### üõ∞Ô∏è Ejemplo (intuitivo)

Imagina un componente de avi√≥n fijado a un shaker:

- Se aplica una excitaci√≥n senoidal de 0.5 $g$ RMS entre 10 y 2000 Hz.

- Se registran tres picos a 120 Hz, 540 Hz y 1370 Hz.

- Esos son modos de vibraci√≥n.

- Alrededor de 540 Hz la amplitud de respuesta es muy alta ‚Üí posible problema de resonancia en servicio.

- Se podr√≠a a√±adir amortiguamiento o modificar rigidez para desplazar esa frecuencia.

Aqu√≠ tienes la simulaci√≥n del an√°lisis seno para un sistema masa‚Äìresorte‚Äìamortiguador:

- <b>Arriba:</b> la amplitud normalizada de la respuesta frente a la frecuencia.

    - El pico indica la frecuencia natural del sistema (~10 Hz en este ejemplo).

    - La amplitud crece fuertemente en resonancia.

- <b>Abajo:</b> la fase entre la fuerza de excitaci√≥n y la respuesta.

    - Antes de la resonancia, la respuesta est√° en fase (‚âà 0¬∞).

    - En resonancia, la fase se retrasa 90¬∞.

    - Despu√©s de la resonancia, la respuesta est√° casi opuesta (‚âà 180¬∞).

Esto reproduce exactamente lo que se observa en un ensayo senoidal real: una clara resonancia con desplazamiento de fase caracter√≠stico y dependencia de la amplitud con la frecuencia.

````python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros del sistema masa-resorte-amortiguador
m = 1.0      # masa (kg)
k = 4000.0   # rigidez (N/m)
c = 20.0     # amortiguamiento (N¬∑s/m)

# Frecuencias de excitaci√≥n (rad/s)
freq_hz = np.linspace(1, 100, 2000)  # de 1 a 100 Hz
omega = 2 * np.pi * freq_hz

# Excitaci√≥n senoidal (amplitud = 1 N)
F0 = 1.0

# Funci√≥n de transferencia (amplitud de respuesta en desplazamiento)
H = 1 / np.sqrt((k - m * omega**2)**2 + (c * omega)**2)
X = F0 * H

# C√°lculo de la fase (rad)
phi = np.arctan2(c * omega, (k - m * omega**2))

# Normalizaci√≥n de amplitud para graficar
X_norm = X / np.max(X)

# Gr√°ficas
plt.figure(figsize=(10, 6))

# Amplitud
plt.subplot(2, 1, 1)
plt.plot(freq_hz, X_norm, label="Amplitud normalizada")
plt.axvline(np.sqrt(k/m)/(2*np.pi), color='r', linestyle='--', label="f‚Çô (frecuencia natural)")
plt.title("Respuesta de un sistema masa‚Äìresorte‚Äìamortiguador ante excitaci√≥n senoidal")
plt.ylabel("Amplitud (normalizada)")
plt.legend()
plt.grid(True)

# Fase
plt.subplot(2, 1, 2)
plt.plot(freq_hz, np.degrees(phi), label="Fase")
plt.axvline(np.sqrt(k/m)/(2*np.pi), color='r', linestyle='--')
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Fase (¬∞)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

### 1. F√≥rmulas clave

- Cociente de amortiguamiento te√≥rico (SDOF viscoso): $$Œ∂ = \frac{‚Äãc‚Äã}{2 m œâ_{n}}$$
- Calidad ($Q$): $Q = \frac{1}{2ùúÅ}$
- M√©todo experimental (half-power / -3 dB): si $f_{1}$ y $f_{2}$ son las frecuencias donde la amplitud cae a $1/2\sqrt{2}$ del pico, entonces: $$ Œ∂ ‚âà \frac{‚Äãf_{2‚Äã}‚àíf_{1}}{2f_{peak}}‚Äã‚Äã $$
- Otra forma (si tienes un registro de decaimiento libre en el tiempo) es el log-decrement:
$$ Œ¥ = \ln{(\frac{‚Äãx_{n}‚Äã}{x_{n+1}})‚Äã} $$ $$ Œ∂ ‚âà \frac{Œ¥}{2œÄ}‚Äã $$

### 2. Qu√© hice en la simulaci√≥n

- Sistema: $ùëö=1kg$, $ùëò=4000N/m$ ‚Üí $ùëì_{ùëõ}‚âà10.07Hz$.
- Tres amortiguamientos: $ùëê=2,20,200 N‚ãÖs/m$
- Barrido de 0.1‚Äì100 Hz, excitaci√≥n de 1 N (FRF de desplazamiento calculada).
- Para cada $c$ calcul√©:
	- $Œ∂$ te√≥rico por la f√≥rmula arriba.
	- Frecuencia de pico y amplitud m√°xima.
	- Frecuencias $f_{1}$ y $f_{2}$ en las que la amplitud cae a $\frac{1}{\sqrt{2}}$ del pico (interpolaci√≥n) y con ellas $ùúÅ$ v√≠a banda -3 dB.
- He mostrado dos gr√°ficas separadas: Amplitud vs frecuencia y Fase vs frecuencia para los tres amortiguamientos.

### 3. Interpretaci√≥n pr√°ctica (qu√© ocurre cuando aumenta Œ∂)

- <b>Pico de resonancia disminuye y se ensancha:</b> a mayor $Œ∂$, menor amplitud m√°xima en la resonancia y mayor ancho de banda. En otras palabras, el pico es menos pronunciado (menos sensible a resonancia).

- <b>Desplazamiento del pico:</b> para amortiguamientos peque√±os el pico est√° muy cerca de $f_{n}$‚Äã; con amortiguamiento mayor el pico puede desplazarse ligeramente y su valor decrece.
- <b>Fase:</b> la transici√≥n de la fase alrededor de la resonancia se hace menos abrupta a mayor amortiguamiento (la pendiente del cambio de fase se suaviza).
- <b>Efecto en servicio:</b> aumentar $Œ∂$ (p. ej. mediante amortiguadores o materiales disipativos) reduce el riesgo de amplitudes extremas en resonancia, pero puede aumentar la respuesta en un rango amplio de frecuencias (dependiendo de la energ√≠a de excitaci√≥n).

### 4. M√©todos pr√°cticos para obtener $Œ∂$ en ensayos

- <b>Banda -3 dB (Sine test / FRF):</b> f√°cil y muy usada cuando tienes una curva clara de FRF (como en el ensayo senoidal). Es preciso si el pico est√° bien definido.

- <b>Log-decrement (ring-down):</b> usa un pulso o excita al sistema y analiza el decaimiento libre. Es muy robusto para determinar $Œ∂$ cuando el sistema puede ser dejado a vibrar libremente.

- <b>Identificaci√≥n modal (curve-fitting):</b> para sistemas con varios modos o acoplamientos se hace un ajuste (curve-fit) de FRFs m√∫ltiples para extraer $f_{n}$, $Œ∂$ y formas modales simult√°neamente.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Par√°metros del sistema masa-resorte-amortiguador
m = 1.0      # masa (kg)
k = 4000.0   # rigidez (N/m)
omega_n = np.sqrt(k / m)
f_n = omega_n / (2 * np.pi)

# Valores de amortiguamiento a comparar
c_values = [2.0, 20.0, 200.0]  # bajo, medio, alto

# Barrido de frecuencias
freq_hz = np.linspace(0.1, 100.0, 4000)
omega = 2 * np.pi * freq_hz
F0 = 1.0  # fuerza de excitaci√≥n (N)

results = []
FRFs = {}
PHIs = {}

for c in c_values:
    # FRF de desplazamiento |X(œâ)|
    H = 1.0 / np.sqrt((k - m * omega**2)**2 + (c * omega)**2)
    X = F0 * H
    FRFs[c] = X
    PHIs[c] = np.arctan2(c * omega, (k - m * omega**2))

    # Œ∂ te√≥rico
    zeta_theoretical = c / (2 * m * omega_n)

    # Pico y -3 dB
    peak_idx = np.argmax(X)
    f_peak = freq_hz[peak_idx]
    peak_mag = X[peak_idx]
    half_power_level = peak_mag / np.sqrt(2)

    # Encontrar cruces -3 dB
    indices_above = np.where(X >= half_power_level)[0]
    if indices_above.size > 0:
        f1_idx = indices_above[0]
        f2_idx = indices_above[-1]
        f1 = freq_hz[f1_idx]
        f2 = freq_hz[f2_idx]
        delta_f = f2 - f1
        zeta_bandwidth = delta_f / (2 * f_peak)
    else:
        f1 = f2 = delta_f = zeta_bandwidth = np.nan

    Q = 1 / (2 * zeta_theoretical)

    results.append({
        "c (N¬∑s/m)": c,
        "Œ∂ theoretical": zeta_theoretical,
        "f_peak (Hz)": f_peak,
        "f1 (-3dB)": f1,
        "f2 (-3dB)": f2,
        "Œ∂ from -3dB": zeta_bandwidth,
        "Q theoretical": Q
    })

df = pd.DataFrame(results)
print("\nResultados del c√°lculo de amortiguamiento:\n")
print(df.round(4))

# Gr√°fica 1: Amplitud
plt.figure(figsize=(10, 5))
for c in c_values:
    plt.plot(freq_hz, FRFs[c], label=f"c={c} N¬∑s/m")
plt.axvline(f_n, linestyle='--', color='r', label=f"f‚Çô ‚âà {f_n:.2f} Hz")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("|X| (m/N)")
plt.title("Amplitud vs frecuencia para distintos amortiguamientos")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Gr√°fica 2: Fase
plt.figure(figsize=(10, 5))
for c in c_values:
    plt.plot(freq_hz, np.degrees(PHIs[c]), label=f"c={c} N¬∑s/m")
plt.axvline(f_n, linestyle='--', color='r')
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Fase (¬∞)")
plt.title("Fase vs frecuencia para distintos amortiguamientos")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

### ‚öôÔ∏è 1. Modelo f√≠sico

Imagina una masa $m$ unida a una base excitada por un resorte de rigidez $k$ y un amortiguador viscoso de coeficiente $c$.
La base (el shaker) se mueve siguiendo una se√±al senoidal:

$$ x_{b}(t) = X_{b} \sin(œât) $$

El desplazamiento de la masa ser√°:

$$ x(t)=X \sin(œât+œï) $$

donde:

- $X_{b}$ ‚Üí amplitud de excitaci√≥n de la base,

- $X$ ‚Üí amplitud de respuesta de la masa,

- $œï$ ‚Üí desfase entre ambas.

### üßÆ 2. Ecuaci√≥n de movimiento

El equilibrio din√°mico de fuerzas sobre la masa es:

$$ m \ddot{x} + c(\dot{x} - \dot{x}_{b}) + k(x - x_{b}) = 0 $$

Si pasamos todo en t√©rminos relativos $z=x-x_{b}$ (movimiento relativo entre masa y base):

$$ m \ddot{z} + c\dot{z} + kz = 0$$

El t√©rmino del lado derecho es la fuerza de excitaci√≥n inercial, generada por la aceleraci√≥n de la base.

### üß© 3. Soluci√≥n en estado estacionario

La excitaci√≥n es senoidal, as√≠ que buscamos una respuesta senoidal del mismo tipo:

$$ x_{b}(t) = X_{b} e^{jœât} $$

$$z(t) = Z e^{jœât} $$

Sustituyendo en la ecuaci√≥n:

$$ ( ‚àí m œâ^{2} + jc œâ + k) Z = m œâ^{2} X_{b} $$

De ah√≠ obtenemos la funci√≥n de transferencia entre desplazamiento de masa y desplazamiento de base:

$$ \frac{Z}{X_{b}} = \frac{m œâ^{2}}{k - m œâ^{2} + jc œâ} $$

o tambi√©n en m√≥dulo:

$$ \lvert \frac{Z}{X_{b}} \rvert = \frac{(œâ/œâ_{n})^2}{\sqrt{(1 - (œâ/œâ_{n})^2)^2 +  (2Œ∂(œâ/œâ_{n}))^2}}


### üìà 4. Barrido senoidal (sine sweep)

Un barrido senoidal consiste en variar la frecuencia $œâ$ de forma controlada, normalmente de forma logar√≠tmica o lineal, y medir la respuesta $\frac{X}{X_{b}}$ o $\frac{X}{F}$ del sistema.

Durante el barrido:

- A bajas frecuencias ($ùúî << ùúî_{n}$): la masa sigue casi igual al movimiento de la base ‚áí $X ‚âà X_{b}$.

- En la resonancia ($ùúî = ùúî_{n}$: el sistema amplifica el movimiento ‚áí $\frac{X}{X_{b}}$ m√°ximo.
