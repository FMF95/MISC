## ⚙️ ¿Qué es un análisis seno (o ensayo senoidal)?

#### Es una prueba de vibración controlada en la que el excitador (shaker) aplica una señal senoidal pura a una estructura, variando la frecuencia a lo largo de un rango definido.
#### El objetivo es observar la respuesta del sistema a esa excitación armónica para identificar sus frecuencias naturales, modos de vibración, y posibles resonancias.

---

### 🔹En ingeniería o física: Análisis senoidal (o análisis de señales senoidales)

Un un “ensayo senoidal” (sine test), una técnica común en ingeniería de vibraciones para caracterizar el comportamiento dinámico de un sistema o componente. Es el estudio del comportamiento de un sistema cuando se somete a una señal senoidal (una onda tipo “seno”).

👉 <b>Usos</b>: análisis de circuitos eléctricos, vibraciones, acústica, estructuras, dinámica, etc.

👉 <b>Ejemplo</b>: aplicar una carga o señal tipo seno a una estructura o sistema y observar su respuesta (amplitud, fase, frecuencia natural, resonancia, etc.).

👉 <b>Objetivo</b>: determinar cómo el sistema responde a estímulos armónicos, lo que ayuda a predecir su comportamiento ante señales más complejas (mediante análisis de Fourier).

---

🔍 Principio básico

Una señal senoidal tiene la forma:

$$ x(t) = A\sin(2𝜋𝑓t +𝜙) $$

donde:

- $A$: amplitud (nivel de vibración),

- $f$: frecuencia,

- $𝜙$ = fase inicial.

Durante el ensayo, se barre la frecuencia $f$ (por ejemplo, de 5 Hz a 2000 Hz), mientras se mantiene una amplitud o nivel de aceleración constante (por ejemplo, 0.5 $g$ RMS).

 ---

### 📈 Tipos de análisis seno

<b>Sine Sweep (barrido senoidal)</b>:
La frecuencia del excitador se incrementa o disminuye lineal o logarítmicamente dentro de un rango.
→ Sirve para detectar resonancias (picos de respuesta).

<b>Sine Dwell</b>:
Se mantiene la excitación a una frecuencia fija (de resonancia) durante un tiempo.
→ Se usa para evaluar la fatiga o el comportamiento estructural en resonancia.

---

### 🧩 Resultados típicos

Del análisis se obtienen curvas de:

- Respuesta en amplitud (aceleración, velocidad o desplazamiento) vs. frecuencia.

- Desfase (fase entre excitación y respuesta).

- Estos datos permiten calcular:

- Frecuencias naturales,

- Factores de amortiguamiento,

- Rigidez dinámica,

- Modos de vibración (si se instrumenta en varios puntos).

---

### 🧪 Aplicaciones comunes

- Ensayos de certificación aeroespacial (MIL-STD-810, DO-160).

- Calibración de modelos de elementos finitos (FEA).

- Validación estructural o de ensamblajes.

- Verificación de equipos electrónicos o mecánicos frente a vibraciones.

---

### 💡 Diferencia con otros tipos de análisis

| Tipo de ensayo         | Señal aplicada               | Uso principal                                                 |
| ---------------------- | ---------------------------- | ------------------------------------------------------------- |
| **Seno (Sine)**        | Onda senoidal pura           | Resonancias y respuesta armónica                              |
| **Aleatorio (Random)** | Ruido con espectro definido  | Simulación de vibraciones reales (como en vuelo o transporte) |
| **Chirp o Sine Sweep** | Seno con frecuencia variable | Identificación modal o búsqueda de resonancias                |
| **Shock (Choque)**     | Pulso transitorio            | Impactos o eventos bruscos                                    |

### ⚙️ 1. Qué información se obtiene de un análisis seno

Cuando realizas un ensayo senoidal (sine test), el objetivo es caracterizar dinámicamente el sistema, es decir, entender cómo responde a excitaciones periódicas.
La información que obtienes se deriva de la relación entre la excitación (entrada) y la respuesta medida (salida).

#### 🔹 Señales medidas

Durante el ensayo, normalmente se miden:

- Aceleración de entrada (sensor en el “shaker” o punto de fijación).

- Aceleraciones de respuesta (uno o varios acelerómetros en la estructura).

A partir de estas señales se obtienen:

| Tipo de dato                                  | Descripción                                                                 | Qué te dice                                                      |
| --------------------------------------------- | --------------------------------------------------------------------------- | ---------------------------------------------------------------- |
| **Amplitud (magnitud)**                       | Aceleración, velocidad o desplazamiento de respuesta                        | Cuánto vibra el sistema a cada frecuencia                        |
| **Fase**                                      | Diferencia de fase entre entrada y salida                                   | Cómo se retrasa o adelanta la respuesta respecto a la excitación |
| **Transmisibilidad (Transfer Function, FRF)** | Cociente entre salida y entrada en frecuencia: $H(f) = \frac{Y(f)}{X(f)}$ | Describe la respuesta dinámica completa                          |
| **Resonancias**                               | Frecuencias donde la amplitud de respuesta es máxima                        | Coinciden con las **frecuencias naturales** del sistema          |
| **Amortiguamiento (ζ)**                       | Se calcula a partir del ancho de banda de un pico resonante                 | Indica la capacidad del sistema para disipar energía             |

---


### 🧠 2. Interpretación de los resultados
#### 📈 a) Curva de amplitud vs. frecuencia

Al hacer un barrido de frecuencia (sine sweep), obtienes una curva como esta:

```txt
Amplitud
│               /\
│              /  \      ← Pico de resonancia
│             /    \
│            /      \
└───────────┬──────────────→ Frecuencia
             f₁
```
- El pico indica una frecuencia natural (f₁) del sistema.

- La amplitud máxima crece cuando el amortiguamiento es bajo.

- Si aplicas distintos puntos de medida, obtienes varios picos → modos de vibración.

#### 📉 b) Curva de fase vs. frecuencia

La fase cambia rápidamente cerca de las resonancias:

- Antes de la resonancia → respuesta en fase con la excitación.

- En resonancia → fase ≈ 90°.

- Después de la resonancia → fase ≈ 180° (respuesta opuesta).

Esto confirma el comportamiento típico de un oscilador amortiguado.

---

### 🔧 3. Cómo se define e introduce la amplitud (RMS, pico, pico a pico)

En vibraciones, la amplitud de excitación puede expresarse de tres formas:

| Tipo                       | Símbolo   | Relación                            | Comentario                   |
| -------------------------- | --------- | ----------------------------------- | ---------------------------- |
| **Pico (peak)**            | $A_{pk}$  | —                                   | Valor máximo instantáneo     |
| **Pico a pico (p-p)**      | $A_{pp}$  | $A_{pp} = 2A_{pk}$                  | De cresta a cresta           |
| **RMS (Root Mean Square)** | $A_{rms}$ | $A_{rms} = \frac{A_{pk}}{\sqrt{2}}$ | Valor eficaz (energía media) |

La amplitud RMS (valor cuadrático medio) representa la energía equivalente de la señal:

$$ A_{rms} ​= \sqrt{ \frac{1}{T} ​∫_{0}^{T} ​[A sin(2πft)]^{2} dt } ​= \frac{A}{\sqrt{​2}}​ $$

👉 Por eso, los niveles de vibración especificados en estándares (como MIL-STD o ISO) se suelen dar en aceleración RMS ($g$ RMS), porque:

- Refleja la energía transmitida, no solo el pico.

- Es más útil para comparar señales o definir la severidad del ensayo.

---

### 📊 4. Cómo se controla la amplitud en el ensayo

En el sistema de control del shaker:

1. Se programa un perfil de frecuencia (por ejemplo, 5 Hz → 2000 Hz en 2 min).

2. Se define un nivel de excitación constante (por ejemplo, 0.5 $g$ RMS).

3. El controlador de vibración ajusta en tiempo real la salida al amplificador del shaker para mantener esa amplitud RMS medida por el acelerómetro de control.

Si el sistema entra en resonancia y la amplitud tiende a subir, el controlador reduce la señal para mantener el nivel RMS deseado — esto evita daños.

---

### 🧩 5. Parámetros derivados importantes

A partir de las curvas de respuesta puedes calcular:

<b>1. Frecuencia natural ($f_{n}$)</b>

Frecuencia en la que ocurre la amplitud máxima.

<b>2. Amortiguamiento ($ζ$ o $Q$)</b>

Del ancho del pico a −3 dB respecto al máximo:
​
$$ ζ = \frac{Δf​}{2fn​} $$

o

$$ Q = \frac{1}{2ζ} ​= \frac{fn​​}{Δf} $$

<b>3. Rigidez dinámica ($k^{*}$)</b>

Relaciona fuerza y desplazamiento:

$$ k^{∗}(ω) = \frac{F(ω)​}{X(ω)} $$


Permite comparar rigidez aparente del sistema en distintas frecuencias.

<b>4. Identificación modal (si hay varios sensores)</b>

Usando las funciones de transferencia entre múltiples puntos, puedes obtener formas modales experimentales y compararlas con modelos FEA.

---

### 🛰️ Ejemplo (intuitivo)

Imagina un componente de avión fijado a un shaker:

- Se aplica una excitación senoidal de 0.5 $g$ RMS entre 10 y 2000 Hz.

- Se registran tres picos a 120 Hz, 540 Hz y 1370 Hz.

- Esos son modos de vibración.

- Alrededor de 540 Hz la amplitud de respuesta es muy alta → posible problema de resonancia en servicio.

- Se podría añadir amortiguamiento o modificar rigidez para desplazar esa frecuencia.

Aquí tienes la simulación del análisis seno para un sistema masa–resorte–amortiguador:

- <b>Arriba:</b> la amplitud normalizada de la respuesta frente a la frecuencia.

    - El pico indica la frecuencia natural del sistema (~10 Hz en este ejemplo).

    - La amplitud crece fuertemente en resonancia.

- <b>Abajo:</b> la fase entre la fuerza de excitación y la respuesta.

    - Antes de la resonancia, la respuesta está en fase (≈ 0°).

    - En resonancia, la fase se retrasa 90°.

    - Después de la resonancia, la respuesta está casi opuesta (≈ 180°).

Esto reproduce exactamente lo que se observa en un ensayo senoidal real: una clara resonancia con desplazamiento de fase característico y dependencia de la amplitud con la frecuencia.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema masa-resorte-amortiguador
m = 1.0      # masa (kg)
k = 4000.0   # rigidez (N/m)
c = 20.0     # amortiguamiento (N·s/m)

# Frecuencias de excitación (rad/s)
freq_hz = np.linspace(1, 100, 2000)  # de 1 a 100 Hz
omega = 2 * np.pi * freq_hz

# Excitación senoidal (amplitud = 1 N)
F0 = 1.0

# Función de transferencia (amplitud de respuesta en desplazamiento)
H = 1 / np.sqrt((k - m * omega**2)**2 + (c * omega)**2)
X = F0 * H

# Cálculo de la fase (rad)
phi = np.arctan2(c * omega, (k - m * omega**2))

# Normalización de amplitud para graficar
X_norm = X / np.max(X)

# Gráficas
plt.figure(figsize=(10, 6))

# Amplitud
plt.subplot(2, 1, 1)
plt.plot(freq_hz, X_norm, label="Amplitud normalizada")
plt.axvline(np.sqrt(k/m)/(2*np.pi), color='r', linestyle='--', label="fₙ (frecuencia natural)")
plt.title("Respuesta de un sistema masa–resorte–amortiguador ante excitación senoidal")
plt.ylabel("Amplitud (normalizada)")
plt.legend()
plt.grid(True)

# Fase
plt.subplot(2, 1, 2)
plt.plot(freq_hz, np.degrees(phi), label="Fase")
plt.axvline(np.sqrt(k/m)/(2*np.pi), color='r', linestyle='--')
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Fase (°)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

He añadido una simulación que compara tres niveles de amortiguamiento y he calculado $ζ$ por dos vías (teórica y por banda -3 dB). Primero, mira la tabla con los resultados (te la he mostrado en la salida). Ahora un resumen interpretativo y las fórmulas importantes:

### 1. Fórmulas clave

- Frecuencia natural: $𝜔_{n} = \sqrt{\frac{k}{m}}$, $f_{n} = \frac{𝜔_{n}}{2π}$
- Cociente de amortiguamiento teórico (SDOF viscoso): $$ζ = \frac{​c​}{2 m ω_{n}}$$
- Calidad ($Q$): $Q = \frac{1}{2𝜁}$
- Método experimental (half-power / -3 dB): si $f_{1}$ y $f_{2}$ son las frecuencias donde la amplitud cae a $1/2\sqrt{2}$ del pico, entonces: $$ ζ ≈ \frac{​f_{2​}−f_{1}}{2f_{peak}}​​ $$
- Otra forma (si tienes un registro de decaimiento libre en el tiempo) es el log-decrement:
$$ δ = \ln{(\frac{​x_{n}​}{x_{n+1}})​} $$ $$ ζ ≈ \frac{δ}{2π}​ $$

### 2. Qué hice en la simulación

- Sistema: $𝑚=1kg$, $𝑘=4000N/m$ → $𝑓_{𝑛}≈10.07Hz$.
- Tres amortiguamientos: $𝑐=2,20,200 N⋅s/m$
- Barrido de 0.1–100 Hz, excitación de 1 N (FRF de desplazamiento calculada).
- Para cada $c$ calculé:
	- $ζ$ teórico por la fórmula arriba.
	- Frecuencia de pico y amplitud máxima.
	- Frecuencias $f_{1}$ y $f_{2}$ en las que la amplitud cae a $\frac{1}{\sqrt{2}}$ del pico (interpolación) y con ellas $𝜁$ vía banda -3 dB.
- He mostrado dos gráficas separadas: Amplitud vs frecuencia y Fase vs frecuencia para los tres amortiguamientos.

### 3. Interpretación práctica (qué ocurre cuando aumenta ζ)

- <b>Pico de resonancia disminuye y se ensancha:</b> a mayor $ζ$, menor amplitud máxima en la resonancia y mayor ancho de banda. En otras palabras, el pico es menos pronunciado (menos sensible a resonancia).

- <b>Desplazamiento del pico:</b> para amortiguamientos pequeños el pico está muy cerca de $f_{n}$​; con amortiguamiento mayor el pico puede desplazarse ligeramente y su valor decrece.
- <b>Fase:</b> la transición de la fase alrededor de la resonancia se hace menos abrupta a mayor amortiguamiento (la pendiente del cambio de fase se suaviza).
- <b>Efecto en servicio:</b> aumentar $ζ$ (p. ej. mediante amortiguadores o materiales disipativos) reduce el riesgo de amplitudes extremas en resonancia, pero puede aumentar la respuesta en un rango amplio de frecuencias (dependiendo de la energía de excitación).

### 4. Métodos prácticos para obtener $ζ$ en ensayos

- <b>Banda -3 dB (Sine test / FRF):</b> fácil y muy usada cuando tienes una curva clara de FRF (como en el ensayo senoidal). Es preciso si el pico está bien definido.

- <b>Log-decrement (ring-down):</b> usa un pulso o excita al sistema y analiza el decaimiento libre. Es muy robusto para determinar $ζ$ cuando el sistema puede ser dejado a vibrar libremente.

- <b>Identificación modal (curve-fitting):</b> para sistemas con varios modos o acoplamientos se hace un ajuste (curve-fit) de FRFs múltiples para extraer $f_{n}$, $ζ$ y formas modales simultáneamente.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Parámetros del sistema masa-resorte-amortiguador
m = 1.0      # masa (kg)
k = 4000.0   # rigidez (N/m)
omega_n = np.sqrt(k / m)
f_n = omega_n / (2 * np.pi)

# Valores de amortiguamiento a comparar
c_values = [2.0, 20.0, 200.0]  # bajo, medio, alto

# Barrido de frecuencias
freq_hz = np.linspace(0.1, 100.0, 4000)
omega = 2 * np.pi * freq_hz
F0 = 1.0  # fuerza de excitación (N)

results = []
FRFs = {}
PHIs = {}

for c in c_values:
    # FRF de desplazamiento |X(ω)|
    H = 1.0 / np.sqrt((k - m * omega**2)**2 + (c * omega)**2)
    X = F0 * H
    FRFs[c] = X
    PHIs[c] = np.arctan2(c * omega, (k - m * omega**2))

    # ζ teórico
    zeta_theoretical = c / (2 * m * omega_n)

    # Pico y -3 dB
    peak_idx = np.argmax(X)
    f_peak = freq_hz[peak_idx]
    peak_mag = X[peak_idx]
    half_power_level = peak_mag / np.sqrt(2)

    # Encontrar cruces -3 dB
    indices_above = np.where(X >= half_power_level)[0]
    if indices_above.size > 0:
        f1_idx = indices_above[0]
        f2_idx = indices_above[-1]
        f1 = freq_hz[f1_idx]
        f2 = freq_hz[f2_idx]
        delta_f = f2 - f1
        zeta_bandwidth = delta_f / (2 * f_peak)
    else:
        f1 = f2 = delta_f = zeta_bandwidth = np.nan

    Q = 1 / (2 * zeta_theoretical)

    results.append({
        "c (N·s/m)": c,
        "ζ theoretical": zeta_theoretical,
        "f_peak (Hz)": f_peak,
        "f1 (-3dB)": f1,
        "f2 (-3dB)": f2,
        "ζ from -3dB": zeta_bandwidth,
        "Q theoretical": Q
    })

df = pd.DataFrame(results)
print("\nResultados del cálculo de amortiguamiento:\n")
print(df.round(4))

# Gráfica 1: Amplitud
plt.figure(figsize=(10, 5))
for c in c_values:
    plt.plot(freq_hz, FRFs[c], label=f"c={c} N·s/m")
plt.axvline(f_n, linestyle='--', color='r', label=f"fₙ ≈ {f_n:.2f} Hz")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("|X| (m/N)")
plt.title("Amplitud vs frecuencia para distintos amortiguamientos")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfica 2: Fase
plt.figure(figsize=(10, 5))
for c in c_values:
    plt.plot(freq_hz, np.degrees(PHIs[c]), label=f"c={c} N·s/m")
plt.axvline(f_n, linestyle='--', color='r')
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Fase (°)")
plt.title("Fase vs frecuencia para distintos amortiguamientos")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

### ⚙️ 1. Modelo físico

Imagina una masa $m$ unida a una base excitada por un resorte de rigidez $k$ y un amortiguador viscoso de coeficiente $c$.
La base (el shaker) se mueve siguiendo una señal senoidal:

$$ x_{b}(t) = X_{b} \sin(ωt) $$

El desplazamiento de la masa será:

$$ x(t)=X \sin(ωt+ϕ) $$

donde:

- $X_{b}$ → amplitud de excitación de la base,

- $X$ → amplitud de respuesta de la masa,

- $ϕ$ → desfase entre ambas.

### 🧮 2. Ecuación de movimiento

El equilibrio dinámico de fuerzas sobre la masa es:

$$ m \ddot{x} + c(\dot{x} - \dot{x}_{b}) + k(x - x_{b}) = 0 $$

Si pasamos todo en términos relativos $z=x-x_{b}$ (movimiento relativo entre masa y base):

$$ m \ddot{z} + c\dot{z} + kz = 0$$

El término del lado derecho es la fuerza de excitación inercial, generada por la aceleración de la base.

### 🧩 3. Solución en estado estacionario

La excitación es senoidal, así que buscamos una respuesta senoidal del mismo tipo:

$$ x_{b}(t) = X_{b} e^{jωt} $$

$$z(t) = Z e^{jωt} $$

Sustituyendo en la ecuación:

$$ ( − m ω^{2} + jc ω + k) Z = m ω^{2} X_{b} $$

De ahí obtenemos la función de transferencia entre desplazamiento de masa y desplazamiento de base:

$$ \frac{Z}{X_{b}} = \frac{m ω^{2}}{k - m ω^{2} + jc ω} $$

o también en módulo:

$$ \lvert \frac{Z}{X_{b}} \rvert = \frac{(ω/ω_{n})^2}{\sqrt{(1 - (ω/ω_{n})^2)^2 +  (2ζ(ω/ω_{n}))^2}}


### 📈 4. Barrido senoidal (sine sweep)

Un barrido senoidal consiste en variar la frecuencia $ω$ de forma controlada, normalmente de forma logarítmica o lineal, y medir la respuesta $\frac{X}{X_{b}}$ o $\frac{X}{F}$ del sistema.

Durante el barrido:

- A bajas frecuencias ($𝜔 << 𝜔_{n}$): la masa sigue casi igual al movimiento de la base ⇒ $X ≈ X_{b}$.

- En la resonancia ($𝜔 = 𝜔_{n}$: el sistema amplifica el movimiento ⇒ $\frac{X}{X_{b}}$ máximo.

- A altas frecuencias ($𝜔 >> 𝜔_{n}$): la masa apenas se mueve ⇒ $\frac{X}{X_{b}} → 0$.

### 📊 5. Respuesta típica

Si representas el módulo $\lvert \frac{X}{X_{b}} \rvert$ frente a $\frac{𝜔}{𝜔_{n}}$, obtienes una curva de resonancia con un pico cuya altura depende del amortiguamiento $ζ$:

| Amortiguamiento       | Altura del pico          | Comentario                                    |
| --------------------- | ------------------------ | --------------------------------------------- |
| Bajo ($\zeta < 0.05$) | Pico muy alto y estrecho | Sistema casi sin disipación (muy resonante)   |
| Medio ($\zeta ≈ 0.1$) | Pico moderado            | Respuesta controlada                          |
| Alto ($\zeta > 0.3$)  | Pico bajo y ancho        | Respuesta amortiguada, sin resonancia marcada |

La fase se comporta así:

- 0° a bajas frecuencias,

- 90° en la resonancia,

- 180° a altas frecuencias.

### 🧠 6. Relación con los ensayos reales

En un shaker test (ensayo vibratorio real):

- El shaker mueve la base con una aceleración senoidal controlada.

- El sensor de control mide esa aceleración (input).

- Acelerómetros adicionales miden la respuesta en la masa o componente (output).

- El sistema de control barre la frecuencia manteniendo la amplitud RMS de entrada constante (por ejemplo, 0.5 $g$ RMS).

El resultado es un conjunto de curvas de respuesta en frecuencia (FRF) que muestran:

- los picos de resonancia (frecuencias naturales),

- el ancho de banda de cada pico (relacionado con $ζ$),

- y los modos de vibración si hay varios sensores.

### ⚖️ 7. Interpretación de la energía

El amortiguamiento $c$ disipa energía proporcional a la velocidad relativa $(\dot{x} - \dot{x}_{b})$.

En cada ciclo senoidal, la energía disipada $E_{d}$ es:

$$ E_{d} =2πζkX^{2} $$

Por tanto:

- Si $ζ$ aumenta, más energía se disipa y el sistema responde menos violentamente.

- Si $ζ→0$, la energía apenas se disipa y la amplitud en resonancia puede crecer mucho → riesgo estructural.

### 🧾 8. En la simulación anterior

En tu simulación se usó:

- $m=1kg$

- $k=4000N/m$ → $f_{n} ≈ 10.07Hz$

- $c=[2,20,200]N \cdotp s /m$

Por tanto:

| $c (N·s/m)$ | $ζ$ teórico | $Q (=1/2ζ)$ | Tipo de respuesta                 |

| 2         | 0.005     | 100       | Resonancia aguda y alta           |
| 20        | 0.05      | 10        | Resonancia moderada               |
| 200       | 0.5       | 1         | Resonancia suave, muy amortiguada |

En las gráficas, pudiste ver que al aumentar $c$:

- se ensancha,

- y la fase cambia más suavemente.
